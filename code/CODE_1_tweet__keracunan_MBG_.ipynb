```python
import pandas as pd

# Load dataset
df = pd.read_csv("Keracunan_mbg.csv")

# Lihat nama kolom
print(df.columns)

# Filter tweet dengan minimal 50 komentar
df_filtered = df[df['reply_count'] >= 50]

# Jumlah data setelah filter
print(f"Jumlah tweet setelah filter: {len(df_filtered)}")

# Simpan hasil
df_filtered.to_csv("Keracunan_mbg_min_reply_50.csv", index=False)

# Tampilkan beberapa data
df_filtered.head()
```
