```python

import pandas as pd

# =========================
# BACA 4 FILE CSV
# =========================
tweet_1 = pd.read_csv("Keracunan_mbg.csv")
tweet_2 = pd.read_csv("Keracunan_mbg_2.csv")
komentar_1 = pd.read_csv("hasil_komentar_keracunan_mbg.csv")
komentar_2 = pd.read_csv("hasil_komentar_keracunan_mbg_2.csv")

# =========================
# AMBIL KOLOM TEKS
# =========================
tweet_1 = tweet_1[["full_text"]].rename(columns={"full_text":"text"})
tweet_2 = tweet_2[["full_text"]].rename(columns={"full_text":"text"})
komentar_1 = komentar_1[["full_text"]].rename(columns={"full_text":"text"})
komentar_2 = komentar_2[["full_text"]].rename(columns={"full_text":"text"})

# =========================
# GABUNGKAN SEMUA DATA
# =========================
gabungan = pd.concat(
    [tweet_1, tweet_2, komentar_1, komentar_2],
    ignore_index=True
)

# HAPUS DATA KOSONG
gabungan = gabungan.dropna(subset=["text"])

# HAPUS DUPLIKAT
gabungan = gabungan.drop_duplicates(subset=["text"])

# RESET INDEX
gabungan = gabungan.reset_index(drop=True)

# SIMPAN HASIL
gabungan.to_csv(
    "gabungan_tweet_komentar_text_bersih.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Jumlah data akhir:", len(gabungan))
print("File tersimpan: gabungan_tweet_komentar_text_bersih.csv")

gabungan.head()
```
